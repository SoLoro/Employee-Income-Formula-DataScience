```python
#ПРОВЕРКА
#Формула: (Вредность (балл.)*180+Базовая зарплата по профессии+(if(Разряд>3):Разряд*1950)+(if(Разряд>4):Разряд*310)+(if(Разряд>5):Разряд*60)+(if(Разряд>6):Разряд*40))*((Объём производства/820-1.01)/3+1))+Премии
#Создаю фрейм для расчёта по разработанной формуле и сравнения с табличным доходом
final= pd.DataFrame(columns =["FIO","Доход","Доход по формуле"])
sum=vredn=basicc=razr=valueRazr=0
for index,row in data.iterrows():
    vredn=row["Вредность (балл.)"]*180
    basicc=profit.loc[profit["Профессия"]==row["Профессия"],"Базовая зарплата"]
    razr=row["Разряд"]
    if(razr>3):
        valueRazr+=(razr-3)*1950
        if(razr>4):
            valueRazr+=(razr-4)*310
            if(razr>5):
                valueRazr+=(razr-5)*60
                if(razr>6):
                    valueRazr+=(razr-6)*40
    if row["Объем производства"]>820:
        sum=ceil((vredn+basicc+valueRazr)*(round((row["Объем производства"]/820-1.01)/3,3)+1))
    else:
        sum=ceil((vredn+basicc+valueRazr)*(round((820/820-1)/3,3)+1))
    neww_row = {"FIO":row["FIO"],"Доход":row["Доход"],"Доход по формуле":sum} 
    final = final.append(neww_row,ignore_index=True)
    sum=vredn=basicc=razr=valueRazr=0
final.head(24)
```
